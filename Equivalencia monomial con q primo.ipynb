```python
print('---------------------------------------------------------------------------------------------')
print('PENSANDO')
#qqqq=[23]
qqqq=prime_range(7 , 32)
#qqqq=prime_powers(7,10)
for q in qqqq:
    print('---------------------------------------------------------------------------------------------')
    print('INICIO DEL CASO Fq con q =',q)
    k.<a>=GF(q)
    R.<x>=PolynomialRing(k)
    monicirred2=[] #Polinomios monicos irreducibles de grado 2
    for pol in R.polynomials(2):
        if pol.is_irreducible():
            if list(pol)[2] == 1:
                monicirred2.append(pol)
                #print(pol)
    print('-------------------------------')
    print('Hay',len(monicirred2),'polinomios monicos irreducibles de grado 2')
    K.<x>=FunctionField(k)
    O=K.maximal_order()
    MS = MatrixSpace(k,2)
    IMS=[] #MATRICES INVERTIBLES
    for matriz in MS:
        if matriz.det()!=0:
            IMS.append(matriz)
    matord=[]
    for matriz in IMS:
        n=1
        brk=0
        while brk == 0 :
            if matriz^n == MS.diagonal_matrix((matriz^n).diagonal()):
                matord.append(n)
                brk=1
            else:
                n=n+1
        #print(matriz)
        #print('tiene orden')
        #print(n)
        #print('--------------')
    #print('FIN LAZO FOR')
    N=Set(matord)
    print('-------------------------------')
    print('Las longitudes posibles para codigos sigma ciclicos son',N)
    print('Solamente nos interesa en este programa que las longitudes sean 3 < n <', floor(q/2 +1)+1)
    for n in N:
        if n in range(4, floor(q/2 +1)+1):
            print('-------------------------------')
            print('ESTUDIAREMOS AHORA CODIGOS DE LONGITUD',n,'Y DIVISORES G DE GRADO 2')
            #print(n)
            matricesordencuatro=0 #CANTIDAD DE MATRICES INVERTIBLES DE 2X2 DE ORDEN n (SIN REDUCIR PGL2)
            mat4=[] #MATRICES INVERTIBLES DE 2X2 DE ORDEN n (SIN REDUCIR PGL2)
            for i in range(len(matord)):
                if matord[i] == n :
                    matricesordencuatro= matricesordencuatro + 1
                    mat4.append(IMS[i])
            print('-------------------------------')
            print('Hay',matricesordencuatro,'matrices de orden' ,n, 'sin reducir PGL2')
            MATGEN=[]
            cantlugfijos=[]
            CONTEO=[]
            for B in mat4:
                #print(B)
                Binv=Matrix(k,2,[B[1][1],-B[0][1],-B[1][0],B[0][0]])
                #print(Binv)
                sigma=K.hom((B[0][0]*x+B[0][1])/(B[1][0]*x+B[1][1]))
                #print(sigma)
                #BUSCAMOS LOS LUGARES DE GRADO 2 FIJOS POR SIGMA
                lg2fijos=[]
                for pol in monicirred2:
                    sigmapol=sigma(pol)
                    arriba=sigmapol.numerator()
                    abajo=sigmapol.denominator()
                    testeo=[((pol).numerator()).divides(arriba), ((pol).numerator()).divides(abajo)]
                    if testeo == [True, False] :
                        lg2fijos.append(pol)
                cantlugfijos.append(len(lg2fijos))
                #print('-------------------------------')
                #print('Existen',len(lg2fijos),'lugares de grado 2 fijos por sigma asociado a la matriz')
                #print(B)
                #print('lg2fijos=',lg2fijos)
                #BUSCAMOS LAS ORBITAS DE LONGITUD 4 ASOCIADAS A SIGMA
                orbitas=[]
                for i in k:
                    brk=0
                    for orb in orbitas:
                        if K.places_finite()[-i] in orb:
                            brk=brk+1
                    if brk == 0:
                        orbita=[K.places_finite()[-i]]
                        aux=[i]
                        brk2=0
                        while brk2 == 0 and len(orbita)<n :
                            num=(Binv[0][0]*aux[len(aux)-1] + Binv[0][1])
                            deno=(Binv[1][0]*aux[len(aux)-1] + Binv[1][1])
                            if deno != 0:
                                new=num/deno
                                aux.append(new)
                                orbita.append(K.places_finite()[-new])
                            else:
                                orbita.append(K.places()[0])
                                brk2=1
                        if len(orbita) == n:
                            #print(orbita)
                            orbitas.append(orbita)
                #CONSTRUIR LOS CODIGOS C_l(ORBITA,g)
                for Q in lg2fijos:
                    #print(Q)
                    #print(Q in R , Q in monicirred2 , Q in K)
                    I=O.ideal(Q)
                    G=I.divisor_of_zeros()
                    #print(G)
                    Base=(G).basis_function_space()
                    #print('La base del espacio de Riemann-Roch asociado a G=',Q,'es')
                    #print(Base)
                    Matgen=[]
                    for orbita in orbitas:
                        lista=[]
                        for base in Base:
                            for orb in orbita:
                                lista.append((base).evaluate(orb))
                        #print(lista)
                        matriz=Matrix(k,len(Base),len(orbita),lista)
                        #print(matriz)
                        Matgen.append(matriz)
                    Cod=[]
                    for matgen in Matgen:
                        cod=LinearCode(matgen)
                        Cod.append(cod)
                    #print(Cod)
                    MAT=[]
                    for cod in Cod:
                        MAT.append(cod.systematic_generator_matrix())
                        MATGEN.append(cod.systematic_generator_matrix())
                        #print(cod.systematic_generator_matrix())
                    conteo=len(list(Set(MAT)))
                    CONTEO.append(conteo)
                #print('--------------------------------CAMBIAMOS DE MATRIZ B--------------------------------')
    #print('FIN LAZO FOR')
            #print('-------------------------------')
            #print('CONTEO=',Set(CONTEO))
            #print('-------------------------------')
            #print('Las cantidades de lugares fijos de grado 2 posibles son',Set(cantlugfijos), 'al ir variando las matrices B')
            #print('-------------------------------')
            #print('LAS MATRICES GENERADORAS DE CODIGOS DISTINTOS, PERO QUIZAS EQUIVALENTES, SON:')
            #for matgen2 in Set(MATGEN):
                #print('-------------------------------')
                #print(matgen2)
            Theta=list(Set(MATGEN))
            theta=len(Theta)
            print('-------------------------------')
            print('HAY', theta, 'MATRICES GENERADORAS DE CODIGOS DISTINTOS, PERO QUIZAS EQUIVALENTES:')
            if theta != 0:
                noequiv=1
                for mu in Theta:
                    for nu in Theta:
                        C1=codes.LinearCode(mu)
                        C2=codes.LinearCode(nu)
                        if not C1.is_permutation_equivalent(C2):
                            noequiv=noequiv+1
                if noequiv == 1 :
                    print('-------------------------------')
                    print('Existe un UNICO codigo sigma-ciclico, salvo equivalencias, de longitud', n, 'y dimension 3 sobre Fq con q =',q)
                else:
                    print('-------------------------------')
                    print('NO ES CIERTO QUE HAY UN UNICO CODIGO, SALVO EQUIVALENCIAS. DE LOS ESTUDIADOS')
    print('-------------------------------')
    print('FIN DEL CASO Fq con q =',q)
print('---------------------------------------------------------------------------------------------')
print('FIN DEL PROGRAMA')
print('---------------------------------------------------------------------------------------------')
```
